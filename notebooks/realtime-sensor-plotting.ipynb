```python
# get latest sensor readings
def get_sensor_readings(drone_object):
    # mambo.sensors.update()
    drone_object.ask_for_state_update()
    sensors = {}
    sensors = copy.deepcopy(drone_object.sensors.__dict__)
    tmp = {}
    if 'sensors_dict' in sensors:
        tmp = sensors['sensors_dict']
        del sensors['sensors_dict']
    sensors.update(tmp)
    
    #quaternions
    (x, y, z) = drone_object.sensors.quaternion_to_euler_angle(drone_object.sensors.quaternion_w, drone_object.sensors.quaternion_x,
                                                   drone_object.sensors.quaternion_y, drone_object.sensors.quaternion_z)
    sensors['qy'] = y    
    sensors['qx'] = x
    sensors['qz'] = z
    
    ts = datetime.datetime.now()

    return pd.DataFrame(sensors, index=[ts])
```

```python
def blocking_callback(id, get_sensor_readings, mambo, stop):

    period = .005  # in seconds (simulate waiting for new data)
    n_show = 100  # number of points to keep and show
    step = 0
    
    while True:
        
        df = get_sensor_readings(mambo)
#         print(df.T)
        
        for key in ['speed_x', 'speed_y', 'speed_z', 'battery']:
             data[key]['data'].stream({'x': [step], 'y':[df.iloc[-1][key]]}, n_show)
        
        if 'DronePosition_posx' in df.columns:
            data['quaternion']['data'].stream({'x': [df.iloc[-1]['DronePosition_posy']], 'y':[df.iloc[-1]['DronePosition_posx']]}, n_show)
            
        push_notebook(handle=sensor_handle)
        push_notebook(handle=battery_handle)
        push_notebook(handle=quaternion_handle)

        time.sleep(period)
        step +=1

        if stop(): #or (mambo.sensors.flying_state=='landed' and step==25):
            print("exit")
            break
```

```python
# set up 
sensor_plot = figure(plot_width=800, plot_height=400)
battery_plot = figure(plot_width=800, plot_height=400)
quaternion_plot = figure(plot_width=800, plot_height=400)

# sensor
colors = {'speed_x' : 'green', 
          'speed_y' : 'yellow',
          'speed_z' : 'blue',
          'battery' : 'red',
          'quaternion' : 'blue'
         }

data = {}

for key in ['speed_x', 'speed_y', 'speed_z', 'battery']:
    data[key] = {}
    data[key]['data'] = ColumnDataSource(data=dict(x=[0], y=[0]))
    data[key]['line'] = sensor_plot.line("x", "y", source=data[key]['data'], line_color=colors[key])

for key in ['battery']:
    data[key] = {}
    data[key]['data'] = ColumnDataSource(data=dict(x=[0], y=[0]))
    data[key]['line'] = battery_plot.line("x", "y", source=data[key]['data'], line_color=colors[key])

    
for key in ['quaternion']:
    data[key] = {}
    data[key]['data'] = ColumnDataSource(data=dict(x=[0], y=[0]))
    data[key]['line'] = quaternion_plot.scatter("x", "y", source=data[key]['data'], line_color=colors[key])
```
